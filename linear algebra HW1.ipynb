```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

\begin{equation}
\begin{cases}
4x+y=2\\
2x+4y=8
\end{cases}
\end{equation}

```python
x = np.linspace(-20, 20, 2)
y1 = 2-4*x
y2 = (8-2*x)/4
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```
