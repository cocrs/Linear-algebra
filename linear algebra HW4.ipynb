```python
%matplotlib inline

import numpy as np
import sympy as sp
import random as rd
import math
import matplotlib.pyplot as plt
```

```python
# 目標函數
def func(w):
    x = sp.Symbol('x')
    equ = 2*x**5 + 5*x**4 - 1680*x**3 + 5650*x**2
    return equ.subs(x, w)

# 目標函數一階導數
def dfunc(w):
    x = sp.Symbol('x')
    diff = sp.diff(2*x**5 + 5*x**4 - 1680*x**3 + 5650*x**2, x)
    return diff.subs(x, w)

def GD(x_start, df, epochs, lr):
    xs = np.zeros(epochs+1)
    x = x_start
    xs[0] = x
    for i in range(epochs):
    #while x > 0.1:
        dx = df(x)
        # v表示x要改變的幅度
        v = - dx * lr
        x += v
        xs[i+1] = x
    return xs
```

```python
minimum = -30

for s in range(10):
    # 起始權重
    x_start = rd.randint(-30, 30)
    # 執行週期數 (依據不同函數作調整)
    epochs = 100
    # 學習率 (依據不同函數作調整)
    lr = 0.000001
    # 梯度下降法 
    w = GD(x_start, dfunc, epochs, lr=lr)
    # 避免出現無限小
    if(math.isinf(w[epochs]) or math.isnan(w[epochs])):
        continue
    print(w)
    if func(w[epochs]) < func(minimum): 
        minimum = w[epochs]
    
    color = 'r'
    x = np.linspace(-30, 30, 2000)
    y = 2*x**5 + 5*x**4 - 1680*x**3 + 5650*x**2
    plt.plot(x, y, 'b')
    w2 = np.zeros(epochs+1)
    for i in range(epochs+1):
        w2[i] = func(w[i])
    plt.plot(w, w2, c=color, label='lr={}'.format(lr))
    plt.scatter(w, w2, c=color, )
    plt.legend()
    plt.show()
    
print("when x =", minimum,"has the minimum")
```
