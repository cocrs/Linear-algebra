```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = 4*x + 20
```

```python
xy = np.array([x, y])
A = np.array([[2, 4],
              [1, 2]])
```

```python
Axy = A@xy
```

```python
plt.xlim(-20,10)
plt.ylim(-20,10)
plt.plot(x, y)
plt.plot(Axy[0], Axy[1])
```

```python
θ = np.linspace(0, 2*np.pi, 100)
```

```python
r = 4
```

```python
x = r*np.cos(θ)
y = r*np.sin(θ)
```

```python
A = np.array([[3,5],
              [4,1]])
```

```python
xy = np.array([x,y])
Axy = A@xy
```

```python
ax = plt.gca()
ax.set_aspect("equal")

plt.plot(x,y)
plt.plot(Axy[0], Axy[1])
```
